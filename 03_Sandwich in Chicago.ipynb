### **01.시카고 샌드위치 맛집 소개 사이트 접근**

```python
from bs4 import BeautifulSoup 
from urllib.request import urlopen
from urllib.request import Request # HTTP 403 err 방지
```

```python
url_base = 'http://www.chicagomag.com'
url_sub = '/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/'
url = url_base + url_sub

headers = {'User-Agent':'Chrome/66.0.3359.181'} # 해당 site에서 크롤링 접근 차단 한 경우, 우회 경로 설정
req = Request(url, headers=headers)

html = urlopen(req)
soup = BeautifulSoup(html, "html.parser")

soup
```

### **02.웹페이지에서 원하는 데이터 추출&정리**

```python
tmp_one = soup.find_all('div', 'sammy')[0] # 첫번째 내용 변수 저장
```

```python
tmp_one.find(class_='sammyRank').get_text() # 랭킹 정보 - 1위
tmp_one.find(class_='sammyListing').get_text() # 그 메뉴 이름과 가게 이름
tmp_one.find('a')['href'] # 연결된 웹 주소 저장
```

**순위, 메인 메뉴, 카페 이름, url 주소 저장**

```python
from urllib.parse import urljoin
```

```python
rank = []
main_menu = []
cafe_name = []
url_add = []

list_soup = soup.find_all('div', 'sammy')

for item in list_soup:
    rank.append(item.find(class_='sammyRank').get_text())
    
    tmp_string = item.find(class_='sammyListing').get_text()

    main_menu.append(re.split(('\n|\r\n'), tmp_string)[0])
    cafe_name.append(re.split(('\n|\r\n'), tmp_string)[1])
    
    url_add.append(urljoin(url_base, item.find('a')['href']))
```

### **05. 정보 50 전부 가져오기**
**tqdm을 적용하여 50페이지 핸들링**  
: Jupyter Notebook에서 상태 진행바를 쉽게 만들어주는 tqdm 모듈  
- 상태 진행바까지 적용하고 다시 샌드위치페이지 50개에 접근  

```python
from tqdm import tqdm_notebook

price = []
address = []

for n in tqdm_notebook(df.index):
    url = df['URL'][n]
    req = Request(url, headers=headers)
    html = urlopen(req)
    soup_tmp = BeautifulSoup(html, 'lxml')
    
    gettings = soup_tmp.find('p', 'addy').get_text()
    
    price.append(gettings.split()[0][:-1])
    address.append(' '.join(gettings.split()[1:-2]))
```

```python
len(price), len(address), len(df) # 50개 전부 저장됐는지 확인
```

```python
df.head(10) # list 10 확인
```

**data에 price&address 추가**

```python
df['Price'] = price
df['Address'] = address

df = df.loc[:, ['Rank', 'Cafe', 'Menu', 'Price', 'Address']]
df.set_index('Rank', inplace=True)
df.head()
```

```python
# file csv로 저장
df.to_csv('03. best_sandwiches_list_chicago2.csv', sep=',', encoding='UTF-8')
```

### **06.맛집 위치 지도에 표기**

```python
import folium
import pandas as pd
import googlemaps
import numpy as np
```

```python
df = pd.read_csv('03. best_sandwiches_list_chicago2.csv', index_col=0)
```

```python
gmaps_key = " ******************** " # 자신의 key를 입력
gmaps = googlemaps.Client(key=gmaps_key)
```

**위도, 경도 정보 받아오기**

```python
lat = []
lng = []

for n in tqdm_notebook(df.index):
    if df['Address'][n] != 'Multiple':
        target_name = df['Address'][n]+', '+'Chicago'
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
        
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
len(lat), len(lng) # 50개 전부 저장됐는지 확인
```

```python
df['lat'] = lat
df['lng'] = lng
df.head()
```

```python
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], zoom_start=11)
folium.Marker([df['lat'].mean(), df['lng'].mean()], popup='center').add_to(mapping) # 중간값으로 map zoom -> 보기 편함
mapping
```

**list 50 전부 map에 표기**

```python
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], zoom_start=11)

for n in df.index:
    if df['Address'][n] != 'Multiple':
        folium.Marker([df['lat'][n], df['lng'][n]], popup=df['Cafe'][n]).add_to(mapping)

mapping
```
